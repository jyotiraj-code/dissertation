## Portfolio Optimization

Maximize returns minimizing risks.
How?
- Strategically allocating the proportion of each asset in the portfolio to optimize the risk-return tradeoff


Markowitz Model: One such portfolio optimization method

How it is implemented?
The expected maximum yield:

$R_i = \sum_{t=1}^{∞}d_{it}r_{it}$

- Here, $r_{it}$ indicates: Anticipated return at a time t per stock invested in
- $d_{it}$: Rate at which return in the ith security

Standard Deviation here: Used as an indicator of the uncertainity or the risk

What happens when we use individual assets within the portfolio?

- $R_{p} = \sum_{t = 1}^{∞}\sum_{i = 0}^{N}d_{it}r_{it}X_{i}$
- $= \sum_{t = 1}^{∞}X_{i}(\sum_{i = 0}^{N}d_{it}r_{it})$
- $= \sum_{i = 0}^{N}X_{i}R_{i}$

The simplified version of the variance of the protfolio can be written as:

$$
\sigma^2_p = \sum_{i=0}^{N} X_i^2 \sigma_i^2 + \sum_{i=0}^{N} \sum_{j=0}^{N} X_i X_j \sigma_{ij}
$$

Here, $X_{i}$ is the percentage of the asset allocated to the ith asset
